---
title: Querying feature item categories
description: How to query metadata on a particular feature item
---

Some of the features (area types and dimensions) for a population type may be
unfamiliar or difficult to understand. This guide shows you how to get metadata
on an item from a feature.

To do this, pass your population type, feature, and item to the
`CensusAPI.query_categories()` method:

```python
from census21api import CensusAPI

api = CensusAPI()

population_type = "UR_HH"
feature = "dimensions"
item = "health_in_general"

api.query_categories(population_type, feature, item)
```

You can use the same method to query category metadata for area types:

```python
feature = "area-types"
item = "msoa"

api.query_categories(population_type, feature, item)
```
