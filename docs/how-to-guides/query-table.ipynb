To retrieve a table from the Census API, you need three things:

- a **population type** (e.g. household reference persons, usual residents)
- an **area type** (e.g. local authorities, Senedd electoral regions)
- some **dimensions** (e.g. english proficiency, economic activity)

## Retrieving a table

With those selected, we can use the `CensusAPI.query_table()` method to get our
table:

```python
from census21api import CensusAPI

api = CensusAPI()

population_type = "HRP"
area_type = "ctry"
dimensions = ("sex", "hh_tenure_9a")

api.query_table(population_type, area_type, dimensions)
```

## Retrieving labelled data

Sometimes the encoded data can be a bit opaque. If you would prefer to use the
labels for the area type and dimensions, use the `use_id` argument:

```python
api.query_table(population_type, area_type, dimensions, use_id=False)
```
